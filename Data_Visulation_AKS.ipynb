```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#loading iris dataset from sklearn library
iris=snr.load_dataset('iris')
```

visualizing the dataset

```python
#line graph
ax=iris.plot(figsize=(15,8),title='visualization on iris dataset')
ax.set_xlabel('x_axis') #rows
ax.set_ylabel('y_axis') #colums
```

```python
#dropping the categorical variable 'species'
data=iris.drop(['species'],axis=1)
```

```python
import numpy as np
import pandas as pd
```

```python
data_new=pd.read_csv('/content/tips.csv')
```

```python
data_new.head()
```

```python
#Relational plot or relplot
snr.relplot(x='total_bill', y='sex',data=data_new)
```

```python
snr.relplot(x='total_bill', y='tip',hue='smoker',data=data_new) #hue indicates color
```

```python
snr.relplot(x='total_bill', y='tip',hue='sex',data=data_new)
```

```python
snr.relplot(x='total_bill', y='tip',hue='smoker',style='time',data=data_new)
```

```python
snr.relplot(x='total_bill', y='tip',hue='smoker',style='time', col='sex',data=data_new)
```

```python
snr.relplot(x='total_bill', y='tip',hue='smoker',style='sex', col='time',data=data_new)
```

```python
snr.relplot(x='total_bill', y='tip',hue='smoker',style='time', col='sex',size='size',data=data_new)
```

```python
#categorical plot or catplot
snr.catplot(x='total_bill', y='day', data=data_new) #even compare categorical variable also
```

```python
snr.catplot(x='tip', y='time', data=data_new)
```

```python
snr.catplot(x='tip', y='sex', data=data_new)
```

```python
#box plot
snr.catplot(x='day',y='tip', kind='box', hue='sex', data=data_new)
```

```python
#pair plot
snr.pairplot(data_new, x_vars=['tip','sex','time'], y_vars=['total_bill'],size=10)
```

```python
#heatmap - will show the corelation between diffrent features
plt.figure(figsize=(5,5))
snr.heatmap(data_new.corr(),annot=True)
```
